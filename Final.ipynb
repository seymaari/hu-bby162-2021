```python
#Kütüphane Otomasyonu
#Veri Tababı: /content/drive/MyDrive/Colab Notebooks/Final.txt

dosya = "/content/drive/MyDrive/Colab Notebooks/Final.txt"
print("Kütüphane Kataloglama Sistemine Hoş geldiniz")

def materyalListele():
  f = open(dosya, 'r')
  for line in f.readlines():
    print(line)
  f.close()

def materyalKaydet():
  materyalAdi = input("Materyal Adını Giriniz.")
  materyalYazarı = input("Materyal Yazarını Giriniz.")
  materyalYayınevi = input("Materyal Yayınevi Bilgisini Giriniz.")
  materyalBasımtarihi = input("Materyal Basım Tarihini Giriniz.")
  materyalIsbn = input("Materyal ISBN Bilgilerini Giriniz.")
  f = open(dosya, 'a' )
  f.write(materyalAdi + " / " + materyalYazarı + " / " + materyalYayınevi + " / " + materyalBasımtarihi + " / " + materyalIsbn + '\n')
  print("Materyal Kaydı Başarı İle Eklendi!")
  f.close()
  
islemSecme = ["görüntüle", "kaydet", "çıkış"]
while islemSecme:
  girilen_islem = input("Katalogdaki materyal listesini görüntülemek için 'görüntüle'\n"
      "Yeni materyal kaydı eklemek için 'kaydet' görevini yazınız.\n"     
      "Sistemden çıkmak için 'çıkış' görevini yazınız.\n"
      "Yapmak istediğiniz işlem için ilgili görevi yazınız: ")
  
  if girilen_islem == islemSecme[0]:
    print("Kütüphane kataloğuna hoş geldiniz..")
    materyalListele()
  elif girilen_islem == islemSecme[1]:
    print("Kataloğa yeni bir materyal kaydı giriniz.")
    materyalKaydet()
  elif girilen_islem == islemSecme[2]:
    break
```
